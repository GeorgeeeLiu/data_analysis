```python
# 导入各种包
import numpy as np
import sys
import re
import lxml
from bs4 import BeautifulSoup
import requests
from lxml import etree
```

## 本节主要以对香港求职网站JobDB进行网络爬虫，搜寻工作职位信息，并将其信息进行整合用Mapreduce进行简要计算Wordcount。

### 开始爬虫：

```python
keyword = 'data analysis'  # 搜寻关于data analysis的工作，可以随意修改
page = str(1)  # 第一页
r = requests.get('https://hk.jobsdb.com/hk/search-jobs/' + keyword + page)  # 解析网页
if r.status_code == requests.codes.ok: # 檢查狀態碼是否 OK
    print("OK，可以继续操作。")
```

### 通过简单爬虫，获得了关于data analysis的职位名称和URL。

```python
root = etree.HTML(r.text)
result = root.xpath("//h1[@class='FYwKg _1GAuD C6ZIU_1my _8QVx6_1my _3NFar_1my sQuda_1my _3srVf_1my _30vru_1my']")
urls = []
for i in range(0, len(result)):
    url = (result[i].xpath("a/@href"))  # 招聘网址
    title = (result[i].xpath("a/div/text()"))  # 招聘标题
    urls.append(url)  # 提取网站
    print(i+1, '职位名称：',title[0], '\n详细资料：', url[0])
```

对上面得出的网站再进行爬虫找出各个工作的工作描述

```python
JD = []  # 创建job description空集JD
for url in urls:
    r_job = requests.get(url[0])
    root_job = etree.HTML(r_job.text)
    result = root_job.xpath("//div[@data-automation='jobDescription']//text()")
    print(url[0])
    for i in result:
        i = i.replace('\n', '').replace('\xa0', '').replace('\r', '')  # 删掉\n,\xa0,\r
        if not i == '':  # 省略掉为空的结果
            if not i == ' ':  # 省略掉为空格的结果
                print(i)
                JD.append(i)
```

看着有些杂乱，而且信息也太多了，接下来尝试分析将这30个工作描述整合进行Mapreduce。

```python
JDt = ' '
for i in JD:  # 将30个工作描述整合到JDt里
    JDt = JDt + i + ' '
rule = re.compile(r"[^a-z^A-Z]")  # 去掉所有标吊符号，仅仅保留单词
JDtext = rule.sub(' ',JDt)
print(JDtext)  # 看看文本长什么样
```

发现很多不是有效单词<br>
接下来使用**pyenchant**包尝试将非有效单词过滤掉<br>
ref:https://pyenchant.github.io/pyenchant/install.html

```python
import enchant
Dic = enchant.Dict("en_US")
word = ''
for i in JDtext.split():
    if len(i)>1 and Dic.check(i):  # 过滤掉单个字母和错误的非有效单词
        word = word + ' ' + i.lower()
JDfile = open("JD.txt", "w")
JDfile.write(word)  # 将所有word写进JD.txt里
```

## 进行Mapper:

```python
! cat  ./JD.txt  | ./mapper.py -->mapper.txt  # 将JD.txt输入到mapper.py执行mapper程序并输出到mapper.txt
file = open('mapper.txt')
print(file.read())  # 生成键值对
```

## 进行Reducer:
生成reducer.txt，里面出现已经计算好词频。

```python
# Reducer
! cat  ./mapper.txt | sort -k1,1 | ./reducer.py -->reducer.txt  # 将mapper.txt输入到reducer.py执行reducer程序并输出到reducer.txt
file = open('reducer.txt')
print(file.read())
```

# 接下来用Wordcount得到的键值对进行分析

先将reduce.txt转化为csv格式方便分析：

```python
import csv
import pandas as pd
file = open('reducer.txt')
with open("mapreduce.csv","w") as csvfile:  # 创建mapreduce.csv用作数据集分析
    writer = csv.writer(csvfile)
    writer.writerow(["keyword", "value"])
    for line in file.readlines():  # 写入多行
        ll = line.strip('\n')
        word, count = ll.split('\t')
        writer.writerow([word,count])

df = pd.read_csv('mapreduce.csv')
df.head()
```
