```python
# 导入各种包
import pprint
import jsonpath
import json
import pprint


from dewupachong import search_by_keywords_load_more_data 

# 以 aj 1 为关键词在小程序搜索商品
keywords = 'aj 1'
r_kw = search_by_keywords_load_more_data(keywords, 0, 1, 1) # 输入关键词，首页，按销量从高到低排
print('首页一个有', len(r_kw), '个商品。\n')
# print(r_kw)
pprint.pprint(r_kw[0]) # 查看第一个商品的信息
```

```python
for i in r_kw:  #查看关于aj 1的商品名称和其编号，和商品价格
    print('商品:',i['title'],'\n商品编号',i['spuId'],'\n价钱：',int(i['price'])/100,'元')
```

```python
# 取第一条Air Jordan 1 Mid Shattered Backboard作为样本，商品编号为32581，查看其最近购买该商品的信息
# 查看最近购买信息
from dewupachong import recensales_load_more_data

r_buyer = recensales_load_more_data(32581)
# pprint.pprint(r_buyer)
for i in r_buyer:
    print('用户：',i['userName'],'\n码数：',i['propertiesValues'],'\n价钱：',int(i['price'])/100,'元','\n购买时间：',i['formatTime'])
```

```python
# 查看商品详细信息
from dewupachong import product_detail_data

r_detail = product_detail_data(spuId=32581, propertyValueId=0)
# pprint.pprint(r_detail)
print(r_detail.keys()) # 查看商品细节有什么信息
```

```python
# 查看商品基本属性
pprint.pprint(r_detail['baseProperties'])
```

```python
# 查看商品销售属性
pprint.pprint(r_detail['saleProperties']['list'])
```
